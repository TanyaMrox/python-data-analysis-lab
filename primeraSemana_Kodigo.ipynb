# Explicación Paso a Paso del Código en Python

## 1. Crear una Lista

```python
ventas_semana = [120, 150, 90, 200, 210]
```

### ¿Qué es una lista?
Una **lista** en Python es una estructura de datos que permite almacenar múltiples valores en una sola variable.

### Características:
- Los elementos están ordenados.
- Se pueden modificar.
- Cada elemento tiene una posición llamada **índice**.
- El índice comienza desde `0`.

### Representación de índices:

| Índice | Valor |
|--------|-------|
| 0      | 120   |
| 1      | 150   |
| 2      | 90    |
| 3      | 200   |
| 4      | 210   |

---

## 2. Crear un Diccionario

```python
producto = {
    "id": 101,
    "nombre": "Laptop Pro",
    "precio": 1200
}
```

### ¿Qué es un diccionario?
Un **diccionario** en Python almacena información en formato:

```python
clave: valor
```

### En este ejemplo:

| Clave     | Valor          |
|-----------|----------------|
| "id"      | `101`          |
| "nombre"  | `"Laptop Pro"` |
| "precio"  | `1200`         |

### Características:
- Permite organizar información relacionada.
- Cada clave debe ser única.
- Se utiliza mucho para representar objetos o datos estructurados.

---

## 3. Acceder a un Elemento de la Lista

```python
print(f"Venta del miércoles: {ventas_semana[2]}")
```

### Resultado:
```python
90
```

### ¿Por qué?
Porque el índice comienza desde `0`:

| Índice | Día        | Valor |
|--------|------------|-------|
| 0      | Lunes      | 120   |
| 1      | Martes     | 150   |
| 2      | Miércoles  | 90    |

---

## 4. Acceder a un Valor del Diccionario

```python
print(f"Precio del producto: {producto['precio']}")
```

### Explicación:
Aquí accedemos al valor asociado a la clave `"precio"`.

```python
producto['precio']
```

### Resultado:
```python
1200
```

### Importante:
En los diccionarios no se usan índices numéricos, sino claves.

---

## 5. Uso de f-strings

```python
f"Texto {variable}"
```

### ¿Qué es un f-string?
Es una forma moderna y sencilla de insertar variables dentro de un texto.

### Ejemplo:

```python
nombre = "Tanya"

print(nombre)
print(f"Hola, mi nombre es {nombre}")
```

### Resultado:
```python
Tanya
Hola, mi nombre es Tanya
```

# Control de Flujo (Condicionales y Bucles)

````python
ventas_semana = [120,150,90,200,210]

for venta in ventas_semana:
    if venta > 150:
        print(f"Venta alta: {venta}")
    else:
        print(f"Venta normal: {venta}")
````

# Explicación del Código: Condicionales y Bucles en Python

## Código original

```python
# Control de Flujo (Condicionales y Bucles)
ventas_semana = [120, 150, 90, 200, 210]

for venta in ventas_semana:
    if venta > 150:
        print(f"Venta alta: {venta}")
    else:
        print(f"Venta normal: {venta}")
```

---

# ¿Qué es el control de flujo?

El **control de flujo** es la forma en la que Python decide:

- Qué instrucciones ejecutar
- Cuándo ejecutarlas
- Cuántas veces hacerlo

En este ejemplo usamos:

- Un **bucle `for`** → para repetir acciones.
- Una **condición `if`** → para tomar decisiones.

---

# 1. La lista

```python
ventas_semana = [120, 150, 90, 200, 210]
```

Aquí estamos creando una **lista**.  
Una lista sirve para guardar múltiples valores dentro de una sola variable.

## Contenido de la lista

| Posición | Valor |
|----------|-------|
| 0        | 120   |
| 1        | 150   |
| 2        | 90    |
| 3        | 200   |
| 4        | 210   |

---

# ¿Qué representa esta lista?

La variable `ventas_semana` almacena las ventas realizadas durante varios días.

Cada número representa una venta individual.

Por ejemplo:

- `120` → venta normal
- `200` → venta alta

---

# 2. El bucle `for`

```python
for venta in ventas_semana:
```

El bucle `for` permite recorrer automáticamente cada elemento de la lista.

Python toma cada valor de `ventas_semana` y lo guarda temporalmente en la variable `venta`.

---

# ¿Cómo funciona el recorrido?

## Primera iteración

```python
venta = 120
```

## Segunda iteración

```python
venta = 150
```

## Tercera iteración

```python
venta = 90
```

## Cuarta iteración

```python
venta = 200
```

## Quinta iteración

```python
venta = 210
```

---

# Tabla recorrida por el bucle

| Iteración | Valor de `venta` |
|-----------|------------------|
| 1         | 120              |
| 2         | 150              |
| 3         | 90               |
| 4         | 200              |
| 5         | 210              |

---

# ¿Qué hace el `if`?

```python
if venta > 150:
```

El condicional `if` evalúa una condición lógica.

En este caso pregunta:

> ¿La venta es mayor a 150?

---

## Operador de comparación

```python
>
```

El operador `>` significa:

> "Mayor que"

---

# Posibles resultados del `if`

## Si la condición es verdadera (`True`)

Se ejecuta:

```python
print(f"Venta alta: {venta}")
```

---

## Si la condición es falsa (`False`)

Se ejecuta el `else`:

```python
print(f"Venta normal: {venta}")
```

---

# 3. El bloque `else`

```python
else:
```

`else` significa:

> "En cualquier otro caso"

Se ejecuta cuando la condición del `if` no se cumple.

---

# Conceptos importantes utilizados

| Concepto     | Explicación                        |
|--------------|------------------------------------|

| `else`       | Ejecuta alternativa                |

| `f-string`   | Inserta variables en texto         |

````python
import matplotlib.pyplot as plt
import numpy as np

# 1. Datos simulados
meses = np.arange(1, 13)
ventas = np.array([100, 120, 150, 200, 260, 340, 430, 540, 670, 810, 960, 1120])

# 2. Cálculos estadísticos
tasa_crecimiento = np.diff(ventas)
total_acumulado = np.sum(ventas)
promedio_ventas = np.mean(ventas)
mes_max_crecimiento = np.argmax(tasa_crecimiento) + 2
max_crecimiento = np.max(tasa_crecimiento)

# 3. Configuración visual (API Orientada a Objetos)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Análisis de Rendimiento de Ventas', fontsize=14, fontweight='bold')

# Gráfico 1: Evolución Anual
ax1.plot(meses, ventas, marker='o', color='#2E86AB', linewidth=2.5, label='Ventas Mensuales')
ax1.set_title('Tendencia de Ventas', fontsize=12)
ax1.set_xlabel('Mes')
ax1.set_ylabel('Unidades')
ax1.grid(axis='y', linestyle='--', alpha=0.5)
ax1.spines[['top', 'right']].set_visible(False)
ax1.legend()

# Gráfico 2: Crecimiento Intermensual
ax2.bar(meses[1:], tasa_crecimiento, color='#F24236', alpha=0.8, label='Aceleración')
ax2.set_title('Tasa de Cambio (Derivada Numérica)', fontsize=12)
ax2.set_xlabel('Mes')
ax2.set_ylabel('Incremento de Unidades')
ax2.grid(axis='y', linestyle='--', alpha=0.5)
ax2.spines[['top', 'right']].set_visible(False)
ax2.legend()

plt.tight_layout()
plt.show()

# 4. Resumen de Métricas
print("=== RESUMEN DE PROYECTO ===")
print(f"Total acumulado en el año: {total_acumulado:,} unidades")
print(f"Promedio de ventas:        {promedio_ventas:,.2f} unidades")
print(f"Mayor aceleración (Mes {mes_max_crecimiento}): +{max_crecimiento:,} unidades")
````
